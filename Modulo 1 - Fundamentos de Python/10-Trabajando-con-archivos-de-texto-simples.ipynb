### Best Practice ... la forma Paythonista !!
Es una buena práctica usar la declaración **with ...** para abrir archivos, esto te permite no preocuparte por cerrar el archivo después de abrirlo. Aquí está la sintaxis:

````python
with open(ruta + '/test_file.txt') as myfile:
    # Fíjate en la indentación!
    # Veremos este tema de la indentación en profundidad más adelante
    lines = myfile.read()
    
    
# El archivo se cierra automáticamente después de la cláusula with ...
````

````python
print(lines)
````

## <font color='green'>Tarea:</font> Añade nuevas líneas al archivo
Añade don nuevas líneas

Tip:
1. Necesitas abrir el archivo en modo 'a'
```Python
    file_to_add = open('totally_new_file.txt','a')
```
2. Escribe dos líneas
```Pythom
    file_to_add.write('New line')
```
3. Lee el archivo usando **read()**
4. Cierra el archivo usando **close()**

````python
# Método 1:
# Se abre el archivo en modo a
file_to_add = open(ruta + 'totally_new_file.txt','a')

# Se escriben dos lineas en el archivo
file_to_add.write('\nLinea 1: New line')
file_to_add.write('\nLinea 2: Second New Lines')

# Se cierra el archivo para luego abrirlo y poder leerlo
file_to_add.close()

# Se cierra el archivo para luego abrirlo y poder leerlo
file_to_add = open(ruta + 'totally_new_file.txt','r')

# Se lee el archivo file_to_add
print(file_to_add.read())

# Se vuelve a cerrar el archivo
file_to_add.close()
````

````python
# Método 2: Abrir sólo una vez el archivo

# Se abre el archivo en modo a+ para permitir leerlo y escribirlo
file_to_add = open(ruta + 'totally_new_file.txt','a+')

# Se escriben dos lineas con la siguiente sentencia.
file_to_add.write('\nLinea 1: New line')
file_to_add.write('\nLinea 2: Second New Lines')

# Cuando se abre un archivo para agregar, el puntero queda al final.
# Para corregir esto, ponemos el puntero en 0 con seek.
file_to_add.seek(0)

# Leer el archivo
print(f'{file_to_add.read()}')

# Cerrar el archivo
file_to_add.close()

````

## <font color='green'>Tarea:</font> Haz lo mismo anterior pero utilizando la sentencia *with* 
La forma *pythonista* de abrir archivos!!

Tip:
1. Necesitarás usar la sentencia *with open ... * 
```Python
    with open('test_file.txt') as archivo:
```
2. Escribe dos lineas
```Pythom
    archivo.write('Esta vez arbimos el archivo como verdadero pythonista y añadimos una línea')
```
3. Lee el archivo utilizando **read( )**

````python
# Método 1. Abrir archivo para escribirlo y luego para leerlo

# Se utiliza en operador de ambiente with y el modo a
with open(ruta + 'totally_new_file.txt', 'a') as archivo:

  # Se escribe el archivo
  archivo.write('\nLinea 1: Esta vez arbimos el archivo como verdadero pythonista y añadimos una línea')
  archivo.write('\nLinea 2: Esta vez arbimos el archivo como verdadero pythonista y añadimos una línea')

# Se abre de nuevo para poder leerlo
with open(ruta + 'totally_new_file.txt') as archivo:

  print(f'{archivo.read()}')
````

````python
# Método 2. Abrir archivo con solo un with

# Se utiliza with para abrir el archivo, y se utiliza el modo a+ para poder leerlo.
with open(ruta + 'totally_new_file.txt', 'a+') as archivo:

  # Se escribe el archivo
  archivo.write('\nLinea 1: Esta vez arbimos el archivo como verdadero pythonista y añadimos una línea')
  archivo.write('\nLinea 2: Esta vez arbimos el archivo como verdadero pythonista y añadimos una línea')
  
  # Cuando se abre un archivo para agregar, el puntero queda al final.
  # Para corregir esto, ponemos el puntero en 0 con seek.
  archivo.seek(0)
  
  # Leer archivo
  print(f'{archivo.read()}')
  
````
